1 Linha: Crie a classe Linha que tem dois atributos, coordenada1 e coordenada2.
Cada coordenada é uma tupla que carrega duas coordenadas cartesianas (x,y) que
denotam pontos do segmento de reta. Faça métodos que calculem o comprimento
do segmento de reta e sua inclinação.

```python
class Linha:
    def __init__(self, coord1, coord2):
        self.coord1 = coord1
        self.coord2 = coord2
    
    def calculaComprimento (self):
        return (((self.coord1[1]-self.coord1[0])**2 + (self.coord2[1]-self.coord2[0])**2)**(1/2))
    
    def calculaInclinacao (self):
        inclinacao = ((self.coord2[1]-self.coord2[0])/(self.coord1[1]-self.coord1[0]))
        if inclinacao >=0:
            return 'Crescente'
        else:
            return 'Decrescente'
    
l= Linha((1, 2), (1, 3))
l.calculaComprimento()
l.calculaInclinacao()
```

a. A classe Figura deve ter o método area.
b. A classe Circulo é subclasse de Figura.
c. A classe Poligono é subclasse de Figura e deve ter os atributos base e altura .
d. As classes Triangulo, Losango, Retangulo e Quadrado são subclasses de
Poligono. Tente criar mais uma generalização aqui .
e. Os polígonos Retangulo e Quadrado devem ter um meio de calcular sua
diagonal. Use composição para isso.
f. Crie uma classe Geometria com uma lista de Figuras com pelo menos uma
figura de cada e imprima suas áreas.


```python
from abc import ABC, abstractmethod
import math
class Figura(ABC):
    def area ():
        pass
    
    def perimetro(self):
        pass
    

    
class Circulo(Figura):
    def __init__(self, raio=1):
        self.raio = raio
        Figura.__init__(self)
    
    def area(self):
        return math.pi * (self.raio**2)
    
    def perimetro(self):
        return 2 * math.pi * self.raio
        

        
class Poligono(Figura):
    def __init__(self, base= 1, altura= 1):
        self.base = base
        self.altura= altura
        Figura.__init__(self)
        
        

class Diagonal(ABC):
    
    def __init__(self):
        super().__init__()
        
    def diagonal(self):
        return math.sqrt((self.base**2) + (self.altura**2))

    
    
class Triangulo(Poligono):
    
    def __init__(self, base, altura):
        super().__init__(base, altura)
        
    def area(self):
        return (self.base * self.altura) / 2
    
    def perimetro(self):
        pass
        
class Losango(Poligono, Diagonal):
    
    def __init__(self, base, altura):
        Poligono.__init__(self, base, altura)
        Diagonal.__init__(self)
        
    def area(self):
        return (self.base * self.altura) / 2
    
    def perimetro(self):
        return self.base*2 + self.altura*2
        
class Retangulo(Poligono, Diagonal):
    
    def __init__(self, base, altura):
        Poligono.__init__(self, base, altura)
        Diagonal.__init__(self)
        
    def area(self):
        return self.base * self.altura
                         
    def diagonal(self):
        return math.sqrt((self.base**2) + (self.altura**2))
    
    def perimetro(self):
        return self.base*2 + self.altura*2
        
class Quadrado(Poligono):
    
    def __init__(self, base, altura):
        super().__init__(base, altura)
        
    def area(self):
        return self.base * self.altura
    
    def perimetro(self):
        return self.base*2 + self.altura*2
    
    
        
class Geometria:
    def __init__(self):
        self.figuras = [
            Circulo(10),
            Triangulo(5, 10),
            Losango(6, 12),
            Retangulo(16,8),
            Quadrado(4, 4)
        ]
            

geometria = Geometria()

for figura in geometria.figuras:
    print(type(figura))
    print("Área: ", figura.area())
    print("Perímetro: " + str(figura.perimetro()) + "\n")
            



        
```
